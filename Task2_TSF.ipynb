### Exploratory Data Analysis 

```python
# counting the items in different category
sns.countplot(mydata.Category, palette=("dark"))
```

```python
#check the sales in subcategory

sns.catplot(x=mydata['Sub-Category'],y=mydata['Sales'],data=mydata.sort_values('Sub-Category',ascending=False),kind='bar')
plt.title('checking the maximum sales items',fontsize=12)
plt.ylabel('Sales',fontsize=17)
plt.xlabel('Sub-category',fontsize=15)
plt.xticks(rotation ='vertical',fontsize=15)
```

```python
#plotting the profit and segment column

sns.catplot(x=mydata['Segment'],y=mydata['Profit'],data=mydata.sort_values('Segment',ascending=False),kind='bar', height=10, aspect=1)
plt.title('Profit in different Segment',fontsize=12)
plt.ylabel('Profit',fontsize=12)
plt.xlabel('Segment',fontsize=12)
```

```python
#which item is more profitable

sns.catplot(y="Profit",x="Sub-Category",data=mydata.sort_values('Profit',ascending=False),kind="boxen",height=6,aspect=3)
plt.title('profitable items in Superstore',fontsize=30)
plt.ylabel('Profit',fontsize=20)
plt.xlabel('Sub-category',fontsize=20)
plt.xticks(rotation ='vertical',fontsize=20)
```

```python
#which category is having max profit and min profit
region_grp=mydata.groupby(['Region'])
```

```python
region_grp['Category','Profit','Sales','Discount'].max()
```

```python
region_grp['Category','Profit','Sales','Discount'].min()
```
